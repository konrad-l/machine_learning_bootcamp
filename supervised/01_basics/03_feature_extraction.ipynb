```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

### <a name='1'></a> Załadowanie danych

```python
def fetch_financial_data(company='AMZN'):
  """
  This function fetches stock market quotations.
  """
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')

df_raw = fetch_financial_data()
df_raw.head()
```

### <a name='2'></a> Utworzenie kopii danych

```python
df = df_raw.copy()
df = df[:5]
df.info()
# zapisano 5 pierwszych wierszy
```

### <a name='3'></a> Generowanie nowych zmiennych

```python
df['day'] = df.index.day
df['month'] = df.index.month
df['year'] = df.index.year
df
```

### <a name='4'></a> Dyskretyzacja zmiennej ciągłej

```python
# stworzenie nowego df
df = pd.DataFrame({'height': [175., 178.5, 185., 191., 184.5, 183., 168.]})
df
```

```python
# storzenie kolumny, w której zapisany zostanie wyniki w postaci 3 automatycznych 
# przedziałów
df['height_cat'] = pd.cut(x=df.height, bins=3)
df
```

```python
# przedstwienie konkretnych przedziałów
df['height_cat'] = pd.cut(x=df.height, bins=(160, 175, 180, 195))
df
```

```python
df['height_cat'] = pd.cut(x=df.height, bins=(160, 175, 180, 195), 
                          labels=['small', 'medium', 'high'])
df
```

```python
pd.get_dummies(df, drop_first=True, prefix='height')
```

### <a name='5'></a> Ekstrakcja cech

```python
df = pd.DataFrame({'lang': [['PL', 'ENG'], ['GER', 'ENG', 'PL', 'FRA'], ['RUS']]})
df
```

```python
df['lang_number'] = df['lang'].apply(len)
df
```

```python
df['PL_flag'] = df['lang'].apply(lambda x: 1 if 'PL' in x else 0)
df
```
