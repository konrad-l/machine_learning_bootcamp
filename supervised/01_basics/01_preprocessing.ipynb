### <a name='0'></a> Import bibliotek

````python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
````

### <a name='1'></a> Wygenerowanie danych

````python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data)
df_raw
````

### <a name='2'></a> Utworzenie kopii danych

````python
df = df_raw.copy()
df.info()
````

### <a name='3'></a> Zmiana typu danych i wstępna eksploracja



````python
# zmiana danych typu 'object' na 'category'
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype(float)
df.info()
````

````python
df.describe()
````

````python
df.describe().T
````

````python
# zmienne categorical
df.describe(include=['category']).T
````

### <a name='4'></a> LabelEncoder

````python
from sklearn.preprocessing import LabelEncoder

# zamiana wartosci 'categorical' na 'binarną' (0 lub 1)
# LabelEncoder wykorzystywany jest do zmiennej docelowej (nie jest brany 
# do podziału zmiennych z tabeli, które są zbiorem danych)
le = LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
````

````python
# alternatywnie można 2 metody 'fit' i 'transform' wykonac poniższą metodą 
le.fit_transform(df['bought'])
````

````python
# ta metoda przedstawi jakie wartości zostały przypisane, tj. array([0, 1, 2...])
# 'no' = 0, 'yes' = 1, 'kolejny element' = 3
le.classes_
````

````python
# przypisanie wartości do kolumny 'bought'
df['bought'] = le.fit_transform(df['bought'])
df
````

````python
# przywrócenie pierwotnych wartości
df['bought'] = le.inverse_transform(df['bought'])
df
````

### <a name='5'></a> OneHotEncoder

````python
from sklearn.preprocessing import OneHotEncoder

# zamiana wartosci 'categorical' na 'binarną' (0 lub 1)
# przypisanie 'M', 'L', 'XL' do wartości binarnych 0 lub 1

# jeśli 'sparse = False' to przechowuje w pamięcia miejsca tylko te gdzie 
# jest '1' - reszta to będą '0'
encoder = OneHotEncoder(sparse=False)
encoder.fit(df[['size']])
````

````python
encoder.transform(df[['size']])
````

````python
encoder.categories_
````

````python
encoder = OneHotEncoder(drop='first', sparse=False)
# wyrzycenie pierwszej wartosci, żeby uniknąć błędu kodowania zero jedynkowego 
# wartosci categorical
encoder.fit(df[['size']])
encoder.transform(df[['size']])
````

### <a name='7'></a> Standaryzacja - StandardScaler

std() - pandas nieobciążony  
std() - numpy obciążony

````python
print(f"{df['price']}\n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")
````

````python
# standaryzowanie zmiennych wokół "0"
(df['price'] - df['price'].mean()) / df['price'].std()
````

````python
# własna funkcja do standaryzowania wyników
def standardize(x):
  return (x - x.mean()) / x.std()

standardize(df['price'])
````

````python
# funkcja do standaryzacji wyników zaimportowana z biblioteki sklearn
# estymator obciążony wariancji (stąd inne wyniki niż w przypadku pandas)
from sklearn.preprocessing import scale

scale(df['price'])
````

````python
# druga wbudowana funkcja w sklearn 
from sklearn.preprocessing import StandardScaler

# 1. tworzymy instancję dla StandardScaler
# 2. wykorzystanie metody 'fit' w celu dopasowania do naszej kolumny 'price' 
# (oblicza statystyki)
# 3. wykorzystanie metody 'transform', żeby przekształcić wyniki
scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])
````

````python
# standaryzacja wyników dla 2 kolumn jednocześnie
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
````

### <a name='8'></a> Przygotowanie danych do modelu

````python
df = df_raw.copy()
df
````

````python
# 1. Przygotowujemy kolumne 'bought' poprzez utworzenie instancji LabelEncoder
# 2. Standaryzujemy zmienne numeryczne poprzez StandardScaler
# 3. Zamiana zmiennych categorical na binarne poprzez get_dummies 
# (drop_first=True usunie jedną niepotrzebną kolumnę)
le = LabelEncoder()

df['bought'] = le.fit_transform(df['bought'])

scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])

df = pd.get_dummies(df, drop_first=True)
df
````
